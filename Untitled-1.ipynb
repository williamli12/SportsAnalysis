```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('2011-ausopen-points.csv')

print(df.head())
```

```python
df['GameWinner'] = pd.to_numeric(df['GameWinner'], errors='coerce')
```

```python
df['GameWinner'] = pd.to_numeric(df['GameWinner'], errors='coerce')

novak_1148_match_data = df[(df['match_id'] == '2011-ausopen-1148') & (df['GameWinner'] != 0)]

print(novak_1148_match_data.head())
```

```python
novak_1148_filtered_data = df[(df['match_id'] == '2011-ausopen-1148')]

novak_1148_rally_data = novak_1148_filtered_data[['ElapsedTime', 'GameWinner', 'PointNumber', 'PointWinner', 'Rally']]

# Display the first few rows to see the result
novak_1148_rally_data.head()
novak_1148_rally_data.to_csv('novak_1148_rally_data.csv', index=False)
```

```python
##novak_1148_match_data.to_csv('novak_1148_filtered_data.csv', index=False)

plot_data = novak_1148_rally_data.copy()

# Drop rows where 'TimePassed' is NaT (because the first row will have no time difference)
# plot_data = plot_data.dropna(subset=['TimePassed'])

# Convert TimePassed to seconds for easier interpretation
#plot_data['TimePassed'] = plot_data['TimePassed'].dt.total_seconds()

# Set up the plot size
plt.figure(figsize=(10, 6))

# Create a boxplot to visualize time passed for each GameWinner (or a scatterplot)
sns.boxplot(x='PointWinner', y='Rally', data=plot_data)

# Add labels and title
# plt.title('Time Passed vs Game Winner')
plt.xlabel('Winner')
plt.ylabel('Rally')

# Display the plot
plt.show()
```

```python
novak_1148_match_data['ElapsedTime'] = pd.to_timedelta(novak_1148_match_data['ElapsedTime'])

filtered_data = novak_1148_match_data[['ElapsedTime', 'GameWinner']]

filtered_data['TimePassed'] = filtered_data['ElapsedTime'].diff()

filtered_data.head()
##filtered_data.to_csv('novak_1148_filtered_data.csv', index=False)
```

```python
##novak_1148_match_data.to_csv('novak_1148_filtered_data.csv', index=False)

plot_data = filtered_data.copy()

# Drop rows where 'TimePassed' is NaT (because the first row will have no time difference)
plot_data = plot_data.dropna(subset=['TimePassed'])

# Convert TimePassed to seconds for easier interpretation
plot_data['TimePassed'] = plot_data['TimePassed'].dt.total_seconds()

# Set up the plot size
plt.figure(figsize=(10, 6))

# Create a boxplot to visualize time passed for each GameWinner (or a scatterplot)
sns.boxplot(x='GameWinner', y='TimePassed', data=plot_data)

# Add labels and title
plt.title('Time Passed vs Game Winner')
plt.xlabel('Game Winner')
plt.ylabel('Time Passed (seconds)')

# Display the plot
plt.show()
```
